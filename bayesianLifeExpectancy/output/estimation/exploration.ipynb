```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle

PMMH_FILE_PREFIX = "./PMMH_"
REGIONS = pd.read_csv('../../data/UN/countries.csv')
BURNIN_ratio = 0.1

def get_estimate(REGION):
    PMMH_FILE = PMMH_FILE_PREFIX + REGION + ".pkl"
    with open(PMMH_FILE, 'rb') as handle:
        PMMH_CHAINS = pickle.load(handle, )

    return PMMH_CHAINS
```

```python
PMMH_CHAINS = get_estimate('WORLD')
for p in PMMH_CHAINS.theta.dtype.names:
    plt.hist(PMMH_CHAINS.theta[p][int(BURNIN_ratio*PMMH_CHAINS.theta[p].shape[0]):])
    plt.title(p)
    plt.show()
```
