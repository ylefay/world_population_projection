```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

FILENAME = "WPP2022_FERT_F02_FERTILITY_RATES_BY_5-YEAR_AGE_GROUPS_OF_MOTHER.csv"
```

```python
dataframe = pd.read_csv(FILENAME, encoding='utf_8')
dataframe = dataframe.drop(['Variant', 'Location code', 'Notes', 'ISO2 Alpha-code', 'ISO3 Alpha-code', 'SDMX code**', 'Type', 'Parent code'], axis=1).set_index('Index')
dataframe.rename(columns={'Region, subregion, country or area *': 'Region'})
dataframe.dropna(inplace=True)
age_columns = ['10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54']
dataframe = dataframe.astype({age_range: float for age_range in age_columns})
dataframe['TFR'] = dataframe[age_columns].sum(axis=1)
```

```python
France = dataframe[dataframe["Region, subregion, country or area *"]=="France"]
```

```python
plt.plot(France.Year, 5*France['TFR']/1000)
```
