```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
LIST_OF_STAGE_III_COUNTRIES = pd.read_csv("../../../data/UN/countries_in_stage_III.csv")
TFR_OF_STAGE_III_COUNTRIES = pd.read_csv("../../../data/UN/TFR_per_country_stage_III.csv")

PMMH_FILE_PREFIX = "./PMMH_fertility_III_"

REGION_OF_INTEREST = "Mauritius"
TFR_OF_INTEREST = TFR_OF_STAGE_III_COUNTRIES[TFR_OF_STAGE_III_COUNTRIES["Region"] == REGION_OF_INTEREST]

BURNIN_ratio = 0.1

PMMH_FILE = PMMH_FILE_PREFIX + REGION_OF_INTEREST + ".pkl"
with open(PMMH_FILE, "rb") as handle:
    PMMH_CHAINS = pickle.load(handle)
print(PMMH_CHAINS.theta.dtype.names)
```

```python
for p in PMMH_CHAINS.theta.dtype.names:
    plt.hist(PMMH_CHAINS.theta[p][int(BURNIN_ratio*PMMH_CHAINS.theta[p].shape[0]):])
    plt.title(p)
    plt.show()
```
