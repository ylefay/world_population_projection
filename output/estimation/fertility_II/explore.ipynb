```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
LIST_OF_STAGE_II_COUNTRIES = pd.read_csv("../../../data/UN/fertility/countries_in_stage_II.csv")
TFR_OF_STAGE_II_COUNTRIES = pd.read_csv("../../../data/UN/fertility/TFR_per_country_stage_II.csv")

PMMH_FILE_PREFIX = "./PMMH_fertility_II_"

REGION_OF_INTEREST = "WORLD"
TFR_OF_INTEREST = TFR_OF_STAGE_II_COUNTRIES[TFR_OF_STAGE_II_COUNTRIES["Region"] == REGION_OF_INTEREST]

BURNIN_ratio = 0.1

PMMH_FILE = PMMH_FILE_PREFIX + REGION_OF_INTEREST + ".pkl"
with open(PMMH_FILE, "rb") as handle:
    PMMH_CHAINS = pickle.load(handle)
```

```python
for p in PMMH_CHAINS.theta.dtype.names:
    plt.hist(PMMH_CHAINS.theta[p][int(BURNIN_ratio*PMMH_CHAINS.theta[p].shape[0]):])
    plt.title(p)
    plt.show()
```
